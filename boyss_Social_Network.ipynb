```python
import networkx as nx  # Import NetworkX for graph operations
import matplotlib.pyplot as plt  # Import matplotlib for plotting
import numpy as np  # Import numpy for numerical operations
from scipy.stats import poisson  # Import the Poisson distribution from scipy.stats
import collections  # Import collections for counting occurrences

# Degree Distribution and Poisson Distribution
degree_sequence = [d for n, d in combined_graph.degree()]  # Generate a list of node degrees for the combined graph

# Count the occurrences of each degree
degree_counts = collections.Counter(degree_sequence)

# Normalize the frequencies by dividing by the total number of nodes
total_nodes = len(combined_graph.nodes())  # Calculate the total number of nodes in the graph
normalized_counts = {k: v / total_nodes for k, v in degree_counts.items()}  # Normalize degree counts by dividing by total nodes

# Plot the frequency of different node degrees
plt.figure(figsize=(10, 6))  # Set the figure size
plt.bar(normalized_counts.keys(), normalized_counts.values(), width=1, color='red', alpha=0.65, label='Degree Distribution')  # Plot the degree distribution
plt.xlabel("Node Degree")  # Label the x-axis as "Node Degree"
plt.ylabel("Frequency")  # Label the y-axis as "Frequency"
plt.title("Degree Distribution and Poisson Distribution")  # Set the title of the plot

# Calculate the Poisson distribution for comparison
mean_degree = np.mean(degree_sequence)  # Calculate the mean degree of the graph
x_values = np.arange(0, max(normalized_counts.keys()) + 1, 1)  # Create an array of x values
poisson_pmf = poisson.pmf(x_values, mean_degree)  # Calculate the Poisson PMF for the given x values and mean

# Plot the Poisson distribution
plt.plot(x_values, poisson_pmf, 'b--', linewidth=2, label='Poisson Distribution')  # Plot the Poisson distribution

plt.legend()  # Add a legend to the plot
plt.grid(True)  # Add grid lines to the plot
plt.show()  # Display the plot
```

```python
# Create a list of sets where each set contains the nodes belonging to a specific city
empirical_partition = list(groups.values())
```

```python
# Print the empirical partition (list of sets)
empirical_partition
```

```python
# Generate a sequence of partitions using Girvan-Newman algorithm
G_copy2 = combined_graph.copy()
partition_sequence = list(nx.community.girvan_newman(G_copy2))
```

```python
# Print the partition sequence
partition_sequence
```

```python
# Function to compute modularity of a given partition
def compute_modularity(G, partition):
    return nx.community.modularity(G, partition)

# Calculate modularity for each partition in the sequence and select the best one
modularities = [compute_modularity(G_copy2, partition) for partition in partition_sequence]
best_partition_index = modularities.index(max(modularities))
best_partition = partition_sequence[best_partition_index]

# Convert the best partition from tuples of sets to a list of sets
best_partition = [set(c) for c in best_partition]


```

```python
# Print the best partition
best_partition
```

```python
import seaborn as sns

# Number of partitions
num_partitions_actual = len(empirical_partition)
num_partitions_predicted = len(best_partition)

# Generate color maps using seaborn color palette
palette_actual = sns.color_palette("tab20", num_partitions_actual)
palette_predicted = sns.color_palette("tab20", num_partitions_predicted)
colors_actual = np.array(palette_actual)
colors_predicted = np.array(palette_predicted)

# Create dictionaries to map nodes to their respective colors for actual and predicted partitions
node_color_map_actual = {}
for i, partition in enumerate(empirical_partition):
    for node in partition:
        node_color_map_actual[node] = colors_actual[i % len(colors_actual)]

node_color_map_predicted = {}
for i, partition in enumerate(best_partition):
    for node in partition:
        node_color_map_predicted[node] = colors_predicted[i % len(colors_predicted)]

# Assign colors to nodes for actual and predicted partitions
node_colors_actual = [node_color_map_actual[n] for n in G_copy2.nodes()]
node_colors_predicted = [node_color_map_predicted[n] for n in G_copy2.nodes()]

# Draw the graphs side-by-side
fig, axes = plt.subplots(1, 2, figsize=(24, 12))

# Actual Communities
ax = axes[0]
pos = nx.spring_layout(G_copy2)
nx.draw_networkx_edges(G_copy2, pos, alpha=0.5, ax=ax)
nx.draw_networkx_nodes(G_copy2, pos, node_color=node_colors_actual, node_size=100, ax=ax)
nx.draw_networkx_labels(G_copy2, pos, font_size=8, ax=ax)
ax.set_title("Actual Communities")

# Predicted Communities (Girvan-Newman)
ax = axes[1]
nx.draw_networkx_edges(G_copy2, pos, alpha=0.5, ax=ax)
nx.draw_networkx_nodes(G_copy2, pos, node_color=node_colors_predicted, node_size=100, ax=ax)
nx.draw_networkx_labels(G_copy2, pos, font_size=8, ax=ax)
ax.set_title("Predicted Communities (Girvan-Newman)")

# Show the plots
plt.show()
```
