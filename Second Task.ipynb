```python
import networkx as nx
import matplotlib.pyplot as plt

# Set the number of nodes and the number of edges to attach from each new node
n = 100  # Number of nodes
m = 2    # Number of edges to attach from each new node

# Generate the Barabási-Albert model network
G = nx.barabasi_albert_graph(n, m)

# Draw the network
pos = nx.spring_layout(G)  # Layout algorithm for node positioning
nx.draw(G, pos, with_labels=True, node_size=250, alpha=0.6,node_color='red')

# Set the labels and title
plt.xlabel('Node')
plt.ylabel('Node')
plt.title('Barabási-Albert Model Network')

# Show the network
plt.show()
```

```python
import networkx as nx
import random
import itertools

def gnp_random_graph(N, p):
    G = nx.Graph()
    G.add_nodes_from(range(N))
    
    for i, j in itertools.combinations(G.nodes(), 2):
        r = random.random()
        if r < p:
            G.add_edge(i, j)
        # Do nothing if r >= p
        
    return G

G = gnp_random_graph(16, 0.15)
nx.draw(G, with_labels=True)
print('Graph has', G.number_of_edges(), 'edge')
```

```python
def gnm_random_graph(N, M):
    G = nx.Graph()
    G.add_nodes_from(range(N))
    
    possible_edges = itertools.combinations(G.nodes, 2)
    edges_to_add = random.sample(list(possible_edges), M)
    G.add_edges_from(edges_to_add)
    
    return G


G = gnm_random_graph(16, 18)
nx.draw(G)
```
